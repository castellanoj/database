```python
import pandas as pd
from NBEATS import NeuralBeats
from sklearn.metrics import mean_squared_error
from math import sqrt

data = pd.read_csv('https://raw.githubusercontent.com/castellanoj/database/main/ipca.csv', encoding='ISO-8859-1')
data = data.values[0:300]
print(data)

print()
data36 = data[300-36:300]
print(data36)
print()
```

```python
model = NeuralBeats(data=data, forecast_length=12, hidden_layer_units=264 )
model.fit(plot=False, verbose=False, epoch=50)
```

```python
forecast = model.predict(predict_data=data36)

print(forecast)

data = pd.read_csv('infla-numpy-prev1.csv')
toRMSE = data[300:312]

print("dados originais")
print(toRMSE)
print(" --------- ")
print("dados do forecast")
print(forecast)
print("---####------#####------#####")
print(sqrt(mean_squared_error(toRMSE, forecast)))

import numpy as np
naiveDez1999 = np.array([1.35, 1.35, 1.35, 1.35, 1.35, 1.35, 1.35, 1.35, 1.35, 1.35, 1.35, 1.35])

print(sqrt(mean_squared_error(toRMSE, naiveDez1999)))
```
